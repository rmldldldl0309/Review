### `apply` 변환
행 또는 열 단위로  복잡한 처리를 하고자 할 때 사용하는 메서드이다  
`apply` 메서드는 인수로 행 또는 열을 받는 함수를 전달받음

```python
df3 = pd.DataFrame(
    {
        'A':[1, 2, 4, 3, 4],
        'B':[2, 3, 1, 2, 3],
        'C':[1, 5, 2, 4, 4]
    }
)
df3
```

```python
# 열
def getColumnMean(x):
    return x.max() - x.min()

# 함수를 변수로 취급
df3.apply(getColumnMean)

# getColumnMean() : 호출한것
```

```python
# 행 
df3.apply(getColumnMean, axis=1)
```

```python
df3.apply(pd.value_counts)
```

```python
def checkAge(row):
    result = None
    if (row.age >= 20): result = 'adult'
    else: result = 'child'
    return result

titanic.apply(checkAge, axis=1)
```

```python
titanic['adult/child'] = titanic.apply(checkAge, axis=1)
titanic
```

##### 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 `category1` 열을 만들어라. `category1` 카테고리는 다음과 같이 정의된다.

1. 20살이 넘으면 성별을 그대로 사용한다.
2. 20살 미만이면 성별에 관계없이 “child”라고 한다.

```python
def getCategory1(row):
    result = None
    if (row.age >= 20): result = row.sex
    else: result = 'child'
    return result

titanic['category1'] = titanic.apply(getCategory1, axis=1)
titanic
```

### 실수값의 카테고리
`cut()` 메서드와 `qcut()` 메서드로 실수 범위에 따라 카테고리를 지정할 수 있다  
`cut()`: 직접 경계선을 지정하여 카테고리를 만드는 메서드
`qcut()`: 갯수를 동일하게 하여 카테고리를 만드는 메서드

```python
age = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
```

`cut` 메서드에 카테고리할 리스트와 경계리스트, 라벨 리스트를 전달하여 카테고리된 `Categorical` 인스턴스를 반환 받는다 
`Categorical` 인스턴스는 `categories` 속성으로 카테고리 라벨을, `codes` 속성으로 카테고리된 인덱스 리스트를 반환한다

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ['미성년자', '청년', '중년', '장년', '노년']
cats = pd.cut(age, bins, labels=labels)
cats
```

```python
cats.categories
```

```python
cats.codes
```

```python
df4 = pd.DataFrame(age, columns=['age'])
# 문자열로 사용하기 위해서는 astype를 통해 str형태로 바꿔줘야 한다
df4['age_cat'] = pd.cut(age, bins, labels=labels).astype(str)
df4
```

```python
data = np.random.randn(1000)
cats = pd.qcut(data, 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
cats
```
