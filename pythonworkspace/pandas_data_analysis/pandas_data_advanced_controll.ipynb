### `merge`함수를 통한 데이터프레임 병합
`merge()`메서드로 공통 열 혹은 행을 기준으로 두 개의 데이터프레임을 병합 할 수 있음  
병합의 기준이 되는 열 혹은 행을 '키'라고 한다  
  
기본적으로 `merge()`메서드는 **inner join**형태를 가진다  
**outer(full) / left / right join**형태로 변경하고자 한다면 `how`인수에 조인 방식을 지정함  
  
`merge()`메서드로 병합을 하려 한다면 동일한 이름의 열 또는 행이 존재해야한다

```python
df1 = pd.DataFrame(
    {
        '고객번호':[1001, 1002, 1003, 1004, 1005, 1006 , 1007],
        '이름':['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
    }, columns=['고객번호', '이름']
)
df1
```

```python
df2 = pd.DataFrame(
    {
        '고객번호':[1001, 1001, 1005, 1006, 1008, 1001],
        '금액':[10000, 20000, 15000, 5000, 100000, 30000]
    }, columns=['고객번호', '금액']
)
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

만약, 키값에 동일한 데이터가 여러개 존재하는 경우 모든 경우의 수를 표현함

```python
df1 = pd.DataFrame(
    {
        '품좀':['튤립', '튤립', '장미', '장미'],
        '꽃잎길이':[1.4, 1.3, 1.5, 1.3]
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        '품좀':['튤립', '장미', '장미', '무궁화'],
        '꽃잎너비':[0.4, 0.3, 0.5, 0.3]
    }
)
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

병합되는 두 데이터프레임의 동일한 컬럼명이 여러개 존재한다면 모두 키가 되기에  
특정한 컬럼만 키로 사용하기 위해 `on`인수로 지정해주어야 한다

```python
df1 = pd.DataFrame(
    {
        '이름':['춘향', '몽룡', '춘향'],
        '날짜':['20180101', '20180102', '20180102'],
        '데이터':['20000', '30000', '100000']
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        '이름':['춘향', '몽룡'],
        '데이터':['여', '남']
    }
)
df2
```

```python
pd.merge(df1, df2, 'outer')
```

```python
pd.merge(df1, df2, on='이름')
```

키가 되는 기준열의 이름이 서로 다르면 `left_on`, `right_on`인수에 사용할 키 컬럼 이름을 지정

```python
df1 = pd.DataFrame(
    {
        '성명':['춘향', '몽룡', '춘향'],
        '날짜':['20180101', '20180102', '20180102'],
    }
)
df2 = pd.DataFrame(
    {
        '이름':['춘향', '몽룡', '길동'],
        '데이터':['여', '남', '남']
    }
)
df1, df2
```

```python
pd.merge(df1, df2, left_on='성명', right_on='이름')
```

```python
df1 = pd.DataFrame(
    {
        '도시':['서울', '서울', '부산', '부산', '부산'],
        '연도':[2000, 2005, 2010, 2000, 2005],
        '인구':[980, 970, 960, 360, 350]
    }
)
df1
```

### `concat`메서드로 데이터 연결

### 피봇 테이블
**피봇 테이블** : 데이터에 이미 존재하는 기본 열을 행 인덱스와 열 인덱스로 사용하는 테이블    
  
pandas에서 피봇테이블을 만들기 위해서는 `pivot()`메서드를 사용한다  
`pivot()`메서드에는 `index`인수로 행 인덱스로 사용할 열을 지정,  
`columns`인수로 열 인덱스로 사용할 열을 지정, `values`인수로 표현할 데이터를 지정

```python
df1 = pd.DataFrame(
    {
        '도시': ['서울', '서울', '서울', '부산', '부산', '부산', '인천', '인천'],
        '연도': [2000, 2005, 2010, 2000, 2005, 2010, 2005, 2010],
        '인구': [980, 970, 960, 360, 350, 350, 250, 260],
        '지역': ['수도권', '수도권', '수도권', '경상권', '경상권', '경상권', '수도권', '수도권']
    }
)
df1
```

```python
df1.pivot(index='도시', columns='연도', values='인구')
```

행 인덱스나 열 인덱스를 리스트로 전달하면 다중 인덱스 피봇 테이블로 생성할 수 있다

```python
df1.pivot(index=['지역','도시'], columns='연도',values='인구')
```

행 인덱스와 열 인덱스 조건에 만족하는 데이터가 2개 이상 존재한다면 피봇테이블을 생성할 수 없다

### 그룹 분석
데이터가 그룹을 이룰 때 그룹의 특성을 보여주는 분석방법  
그룹 분석은 피봇테이블과 다르게 키에 의해 결정되는 데이터가 여러 개 있을 경우   
미리 지정한 연산을 통해 해당 그룹의 대표값을 계산 

#### `groupby`메서드
`groupby()`메서드는 그룹 별로 분류하여 그룹 객체를 생성  
그룹 객체는 그룹 연산메서드등을 포함하고 있다

#### 그룹 연산 매서드
- `size()`, `count()`: 그룹 데이터의 갯수
- `mean()`,`median()`, `min()`, `max()`: 그룹 데이터의 평균, 중앙값, 최소값, 최대값
- `sum()`, `prod()`, `std()`, `var()`, `quantile()`: 그룹 데이터의 합계, 곱, 표준편차, 분산, 사분위 수
- `first()`, `last()`: 그룹 데이터의 처음 값, 마지막 값  

- `agg()`, `aggregate()`: 그룹 연산 메서드를 직접 생성하여 사용하도록 하는 메서드,  
여러가지 그룹 연산을 동시에 하려할 때 해당 그룹 연산을 리스트로 전달하여 사용하도록 하는 메서드
- `discribe()`: 하나의 대표값이 아니라 여러 개의 값을 데이터프레임으로 구한다
- `apply()`: 하나의 대표값이 아니라 여러 개의 값을 데이터프레임으로 구하는데 원하는 
그룹 연산이 없을 때 메서드를직접 생성하여 사용하는 메서드
- `transform()`: 대표 값을 생성하는 것이 아니라 데이터 자체를 변경하는 메서드

```python
df2 = pd.DataFrame(
    {
        'key1': ['A', 'A', 'B', 'B', 'A'],
        'key2': ['one', 'two', 'one', 'two', 'one'],
        'data1': [1, 2, 3, 4, 5],
        'data2': [10, 20, 30, 40, 50]
    }
)
df2
```

```python
dg = df2.groupby(df2.key1)
dg
```

```python
iris = seaborn.load_dataset('iris')
iris.head()
```

```python
def peak_to_peak_ratio(group_row):
    return group_row.max() / group_row.min()

iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
iris.groupby(iris.species).describe()
```

```python
def top3_petal_length(df):
    return df.sort_values(by = 'petal_length', ascending=False)
 
iris.groupby(iris.species).apply(top3_petal_length)
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=['소', '중', '대']).astype(str)

iris['petal_length_class'] = iris.groupby(iris.species).petal_length.transform(q3cut)
iris[['petal_length', 'petal_length_class']].head(10)
```

### `pivot_tabe`메서드
`pivot_tabe`메서드는 `groupby`메서드 결과를 피봇 테이블롸 결과를 보여주는 메서드  
`pivot_tabe(data, values, index, columns, aggfunc, fill_value,margins, margins_name)`
- `data`: 그룹 분석할 데이트 프레임(데이터프레임의 메서드일 때는 불필요)
- `values`: 분석할 열
- `index`: 행 인덱스로 사용할 열
- `columns`: 열 인덱스로 사용할 열
- `aggfunc`: 분석 메서드
- `fill_value`: NaN을 대체한 값
- `margins`: 데이터 북석 결과를 오른쪽과 아래쪽에 표시할 지 여부
- `margins_name`: 마진 행렬의 이름

```python
df1.pivot_table(values='인구',index='도시', columns='연도', margins=True, margins_name='평균')
```

```python
df1.pivot_table(values='인구',index='도시', columns='연도',aggfunc='sum', margins=True, margins_name='합계')
```

```python
df1.pivot_table(values='인구',index=['지역', '도시'])
```
